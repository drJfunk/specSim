```python
from evo import PulseIntegrator, evo
```

```python
tGrid = linspace(0.,10.,1000.)
tGridfine = linspace(0.,10.,5000.)
```

```python
tmpFine = map(lambda t: PulseIntegrator(t,10.,5000.,[]),tGridfine)
```

```python
tmp = map(lambda t: PulseIntegrator(t,10.,5000.,[]),tGrid)
```

```python
plot(tGrid,tmp)
```

```python
def fast(t,emin,emax):
    
    nE = 500.
    eGrid = linspace(emin,emax,nE)
    dE = diff(eGrid)[0]
    
    
    #dE = (emax-emin)/nE
    tmp = array(map(lambda ene: evo(ene,t,[]),eGrid))
    tmp *= dE
    tmp = tmp.sum()
    
    return tmp
    
    
    
```

```python
from scipy.interpolate import interp1d
```

```python
intP = interp1d(tGrid,tmp)
```

```python
plot(tGridfine,intP(tGridfine),'-')
plot(tGridfine,tmpFine)
xlim(0.2,.8)
```

```python
x=(intP(tGridfine)-tmpFine)/tmpFine
```

```python
plot(tGridfine,x)
xlim(0,1)
```
